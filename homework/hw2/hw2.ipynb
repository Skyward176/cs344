- Compare the medians
- we have the median index -1 smaller numbers. Now compare the number one above te smaller median with the number one below the larger median. repeat. when our count == n, we found the median.

```python
import random

def find_kth(a, b, k):
    """
    Finds the k-th smallest element in two sorted arrays in O(log k) time.
    Note: k is 1-indexed (e.g., k=1 means the 1st smallest element).
    """
    # Ensure 'a' is always the smaller array to optimize bounds checking
    if len(a) > len(b):
        return find_kth(b, a, k)
    
    # Base cases
    if not a:
        return b[k - 1]
    if k == 1:
        return min(a[0], b[0])

    # Compare the (k/2)-th elements of both arrays
    # Use min(len(a), k // 2) to avoid out-of-bounds if 'a' is shorter than k/2
    i = min(len(a), k // 2)
    j = k - i

    if a[i - 1] < b[j - 1]:
        # The first 'i' elements of 'a' cannot contain the k-th smallest element.
        # Discard them and search for the (k - i)-th element in the remainder.
        return find_kth(a[i:], b, k - i)
    else:
        # The first 'j' elements of 'b' cannot contain the k-th smallest element.
        # Discard them and search for the (k - j)-th element in the remainder.
        return find_kth(a, b[j:], k - j)

def find_median(a, b):
    """Problem wrapper: Finds the n-th smallest element of A U B where |A| = |B| = n"""
    n = len(a)
    return find_kth(a, b, n)

def naive_find_median(a, b):
    """A naive O(n log n) implementation to verify our O(log n) algorithm against."""
    combined = sorted(a + b)
    return combined[len(a) - 1]  # n-th smallest is at index n-1

def generate_test_case(n):
    """Generates two sorted arrays of length n containing distinct integers."""
    # 1. Generate a large pool of numbers and pick 2n distinct ones
    pool = random.sample(range(1, n * 10 + 1), 2 * n)
    pool.sort()  # The combined union is now sorted and distinct
    
    # 2. Randomly distribute the 2n elements into A and B so both get exactly n elements
    indices = list(range(2 * n))
    a_indices = set(random.sample(indices, n))
    
    a, b = [], []
    for i in range(2 * n):
        if i in a_indices:
            a.append(pool[i])
        else:
            b.append(pool[i])
            
    return a, b

if __name__ == "__main__":
    # 1. Test the specific example from your problem description
    a_example = [0, 1, 6, 9, 12, 13, 18, 20]
    b_example = [2, 4, 5, 8, 17, 19, 21, 23]
    
    print("--- Testing Problem Example ---")
    print(f"A: {a_example}")
    print(f"B: {b_example}")
    res = find_median(a_example, b_example)
    print(f"Result: {res} (Expected: 9)\n")

    # 2. Run randomized tests to verify correctness
    print("--- Running Randomized Tests ---")
    num_tests = 5
    n_size = 10  # Try changing this to a larger number to see it scale
    
    for test_num in range(1, num_tests + 1):
        a_test, b_test = generate_test_case(n_size)
        expected = naive_find_median(a_test, b_test)
        actual = find_median(a_test, b_test)
        
        status = "PASS" if expected == actual else "FAIL"
        print(f"Test {test_num}: {status} | Expected: {expected}, Actual: {actual}")
        
        if status == "FAIL":
            print(f"  A: {a_test}")
            print(f"  B: {b_test}")
```

```python
# Helper function returns a tuple: (BestOverallProduct, MaxProductEndingAtCurrent)
def MaxProdLinearHelper(A, currentIndex):
    # Base Case: At the very first element, both the overall max and the max 
    # ending at this index are simply the first element itself.
    if currentIndex == 0:
        return (A[0], A[0])

    # Recursive Step: Get the results from the array up to the previous element.
    (bestOverallProduct, maxProductEndingAtPrevious) = MaxProdLinearHelper(A, currentIndex - 1)

    # Calculate the max product that ends exactly at the current element.
    # We have two choices:
    # 1. Start a brand new subarray starting at the current element: A[currentIndex]
    # 2. Extend the previous best suffix by multiplying it: maxProductEndingAtPrevious * A[currentIndex]
    maxProductEndingAtCurrent = max(A[currentIndex], maxProductEndingAtPrevious * A[currentIndex])

    # Determine if the newly found suffix product beats our historical best overall product.
    if maxProductEndingAtCurrent > bestOverallProduct:
        bestOverallProduct = maxProductEndingAtCurrent

    # Return both values up the recursive chain.
    return (bestOverallProduct, maxProductEndingAtCurrent)


# Main wrapper function to kick off the recursion and extract the final answer.
def MaxProd(A):
    lastIndex = len(A) - 1
    (finalMaxProduct, _) = MaxProdLinearHelper(A, lastIndex)
    return finalMaxProduct
A = [3, 0.1, 5, 20, 4]

print(MaxProd(A))
```
